```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
import pymysql
tStart = time.time()

#tag
def get_attr(addr):
    hashtag = []
    ctype = []
    driver.get(addr)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    for ele in soup.select('a.tag-item'):
        #print(ele.text)
        n = ele.get('title').strip()
        hashtag.append(n)
        ctype.append("")
    return hashtag, ctype

#main
driver =  webdriver.Chrome("C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe")
addr = "https://www.taiwan.net.tw/m1.aspx?sNo=0000193"
db = pymysql.connect(
    host="localhost",
    user="root",
    passwd="12345",
    database='homestead'
)
cursor = db.cursor()

tag, ntype = get_attr(addr)
org_id = 0

#加入attr資料到資料庫
for each in range(len(tag)):
    qurey = "SELECT * FROM `site_attr` ORDER BY id DESC LIMIT 1"
    cursor.execute(qurey)
    for row in cursor.fetchall():
        org_id = row[0]
    a_id = org_id+1
    add_data = ("INSERT INTO site_attr"
                "(id,type,tag,color,shape) "
                "VALUES (%s, %s, %s, %s, %s)")
    data = (a_id, ntype[each], tag[each],'red','square')
    cursor.execute(add_data, data)
#     db.commit()
# 博物館重複喔~刪掉就好~

tEnd = time.time()
print("Run Time：", tEnd - tStart)
```
