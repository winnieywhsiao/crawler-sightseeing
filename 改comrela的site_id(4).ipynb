```python
import pymysql
import time

tStart = time.time()

cnx = pymysql.connect(
    host="localhost",
    user="root",
    passwd="12345",
    database='test'
    )

cursor = cnx.cursor()
re_cur = cnx.cursor()
h_cur = cnx.cursor()
cursor2 = cnx.cursor()
```

```python
href = []
path = []
near = "SELECT near_site, city_name, near_href, id FROM hotel_relationship"
cursor2.execute(near)
count = 1

for row in cursor2.fetchall():
    # near
    n = row[0]
    nct = row[1]
    url = row[2]
    c_id = row[3]
    href.append(url)
    
    check = url[31:41]
    if check == "Attraction":
        site = "SELECT id FROM site_data where name = '" + n + "' and city_name = '" + nct + "'"
        cursor.execute(site)
        fet = cursor.fetchall()
        for i in range(len(fet)):
            site_id = fet[i][0]
            # print(site_id)
            path.append(site_id)
            
    elif check == "Restaurant":
        resta = "SELECT id FROM resta_data where name = '" + n + "' and city_name = '" + nct + "'"
        re_cur.execute(resta)
        re_fet = re_cur.fetchall()
        for j in range(len(re_fet)):
            resta_id = re_fet[j][0]
            # print(resta_id)
            path.append(resta_id)
            
    elif check == "Hotel_Revi":
        hotel = "SELECT id FROM hotel_data where name = '" + n + "' and city_name = '" + nct + "'"
        h_cur.execute(hotel)
        h_fet = h_cur.fetchall()
        for i in range(len(h_fet)):
            hotel_id = h_fet[i][0]
            # print(hotel_id)
            path.append(hotel_id)
    
    else:
        # print("None")
        path.append("None")
# print(path)
# print(len(path))
# print(len(href))
```

```python
cmr_cur = cnx.cursor()

for p in range(len(path)):
    num = str(p+1)
    cmr_sid = "UPDATE comment_relationship SET sid = '" + path[p] + "' WHERE id = '" + num + "'"
    # print(cmr_sid)
    
    cmr_cur.execute(cmr_sid)
    # cnx.commit()

tEnd = time.time()
print("Run Time：", tEnd - tStart)
```
